```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
df = pd.read_csv('iris.csv', index_col=0)
df.head()
```

```python
df.info()
```

```python
plt.figure(num = 'a', figsize=(10, 8))
plt.subplot(221)
df.SepalLengthCm.hist()
plt.title('SepalLengthCm')
plt.subplot(222)
df.SepalWidthCm.hist()
plt.title('SepalWidthCm')
plt.subplot(223)
df.PetalLengthCm.hist()
plt.title('PetalLengthCm')
plt.subplot(224)
df.PetalWidthCm.hist()
plt.title('PetalWidthCm');
```

```python
sns.pairplot(data = df, vars= df.columns[:-1], diag_kind=None, hue = 'Species');
```

```python
var = df[df.drop(columns = 'Species').columns].values
```

```python
range_ = np.arange(1, 9)
k_inertia = []

for n in range_:
    kmeans = KMeans(n_clusters=n)
    kmeans.fit(var)
    k_inertia.append(kmeans.inertia_)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 7))
sns.lineplot(x = range_, y = k_inertia, ci = None, marker = 'o', color = 'r')
plt.grid(True)
plt.title('Plot of number of cluster versus kmeans inertia')
plt.xlabel('Number of Clusters')
plt.ylabel('KMeans Inertia');
```

From the graph above we can easy conclude the number of clusters will be that point about which our KMeans Inertia began to become negligible, which is also the point where the curve shoots up. Hence our number of clusters lies between 2 and 3 and 3 will be a more befitting number of clusters
